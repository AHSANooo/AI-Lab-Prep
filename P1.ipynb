```python
heuristics = {
    'A': 10,
    'B': 8,
    'C': 5,
    'D': None,
    'E': 3
}
```

```python
def fixer(heuristics):
    result= []
    for i in heuristics:
        if heuristics[i]==None:
            heuristics[i]=0
        if heuristics[i]<=6:
            result.append(i)
    return result

fixer(heuristics)
```

```python
class Node:

    def __init__(self, name, parent,cost):
        self.name = name
        self.parent = parent
        self.cost = cost
    

    def display_info(self):

        print(f"Name : {self.name} \n Cost {self.cost}")

obj = Node("FKT", None, 10) 
obj.display_info()
```
