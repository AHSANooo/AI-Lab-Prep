```python
import numpy
import sklearn
import pandas
import matplotlib
import seaborn
```

```python
heuristics = {
    'A': 10,
    'B': 8,
    'C': 5,
    'D': None,
    'E': 3
}
```

```python
def fixer(heuristics):
    result= []
    for i in heuristics:
        if heuristics[i]==None:
            heuristics[i]=0
        if heuristics[i]<=6:
            result.append(i)
    return result

fixer(heuristics)
```
